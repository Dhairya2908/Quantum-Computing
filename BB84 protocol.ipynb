```python
%matplotlib inline
from qiskit import *
from qiskit.visualization import *
from qiskit.tools.monitor import *
```

```python
Circuit = QuantumCircuit(1,1)
Circuit.h(0)
Circuit.measure(range(1),range(1))
Circuit.draw(output='mpl')
```

```python
#running on the simulator
#these are random bits generated by Alice
n = 200
backend = Aer.get_backend('qasm_simulator')
job_sim = execute(Circuit,backend,shots=n,memory = True)
result = job_sim.result()
counts = result.get_counts()
print(counts)
bits_alice = [int(q) for q in job_sim.result().get_memory()] 
print(bits_alice)
```

```python
#Alice chosing random basis for the bits she got
job = execute(Circuit,backend,shots=n,memory = True)
basis_alice = [int(q) for q in job.result().get_memory()]
print(basis_alice)
```

```python
#bob generate own random basis
job= execute(Circuit,backend,shots=n,memory=True)
basis_bob = [int(q) for q in job.result().get_memory()]
print(basis_bob)
```

```python
#understand this and comment well!
#now bob need to measure the qubits sent by alice in his basis

# Now, Alice codes each bit of her initial string as a qubit and sends it to Bob, who measures in his basis

bits_bob = []

for i in range(n):
    circ_send = QuantumCircuit(1,1)
    if bits_alice[i]:  # Alice has to send '1'
        circ_send.x(0)

    if basis_alice[i]: # Alice codes in basis |+>, |->
        circ_send.h(0)

        
    # Alice sends the qubit to Bob and he measures
    
    if basis_bob[i]: # Bob has to measure in basis |+>, |->
        circ_send.h(0) 


        
    
    
    circ_send.measure(0,0)
    
    job = execute(circ_send, backend, shots = 1, memory = True)
    bits_bob.append(int(job.result().get_memory()[0]))
    
print(bits_bob)
    
```

```python
#now we need to find out the key.
key = []

for i in range(n):
    if basis_alice[i] == basis_bob[i]:
        key.append(bits_bob[i])
        
print("Key length", len(key))
print(key)
```
